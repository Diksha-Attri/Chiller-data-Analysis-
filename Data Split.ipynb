```python
import os
import pandas as pd

def split_csvs_by_chiller(input_folder, output_folder, chiller_col='ChillerName', chunksize=100_000):
    os.makedirs(output_folder, exist_ok=True)
    chiller_files = {}  # Store file handles to avoid reopening

    # Loop through all CSV files in the folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".csv"):
            file_path = os.path.join(input_folder, filename)
            print(f"üìÇ Processing: {filename}")

            for chunk in pd.read_csv(file_path, chunksize=chunksize):
                chunk = chunk.dropna(subset=[chiller_col])

                for chiller, group in chunk.groupby(chiller_col):
                    safe_name = str(chiller).replace(" ", "_").replace("/", "_")
                    out_path = os.path.join(output_folder, f"{safe_name}.csv")
                    
                    # Append to file if exists, else write header
                    write_header = not os.path.exists(out_path)
                    group.to_csv(out_path, mode='a', index=False, header=write_header)

    print("‚úÖ All files processed and saved by chiller name.")

# Example usage
input_folder = "D:/Dabour chiller/data/Raw"     # Folder with 11 CSV files
output_folder = "D:/Dabour chiller/data/by_chiller"  # Folder to save chiller-wise files
split_csvs_by_chiller(input_folder, output_folder, chiller_col='name.1')
```

```python
import os
import pandas as pd

def count_total_rows(folder_path):
    total_rows = 0
    file_count = 0

    for file_name in os.listdir(folder_path):
        if file_name.endswith(".csv"):
            file_path = os.path.join(folder_path, file_name)
            try:
                row_count = sum(1 for _ in open(file_path)) - 1  # subtract 1 for header
                print(f"{file_name}: {row_count} rows")
                total_rows += row_count
                file_count += 1
            except Exception as e:
                print(f"‚ö†Ô∏è Failed to read {file_name}: {e}")

    print(f"\n‚úÖ Total files: {file_count}")
    print(f"üìä Total rows across all files: {total_rows}")

# Example usage
folder = "D:/Dabour chiller/data/by_chiller"
count_total_rows(folder)
```
